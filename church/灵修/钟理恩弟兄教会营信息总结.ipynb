### 钟理恩弟兄教会营信息总结

弟兄姊妹下午好，

接下来我为大家回忆一下钟理恩弟兄在今年教会营分享的信息。感谢主的带领。若是有遗漏或不妥的地方，希望弟兄姊妹帮助我及时补充和更正。

今年教会营的主题是以大卫为出发点，教我们做一个合乎神心意的人。钟理恩弟兄从以下四个小点来为我们阐明这个主题:
* 1. 我们要做**一个坐在主跟前的人**
* 2. 我们要做**一个专心仰赖主的人**
* 3. 我们要做**一个让主居心中宝座的人**
* 4. 我们要做**一个信心坚固的人**

#### 1. 一个坐在主跟前的人
* 1. 大卫每每在遇事不决和遭受苦难的时候，总是愿意把自己带到耶和华的跟前，恳切地祈祷；他明白主的心意，在他成为以色列的王，有了自己的王国和子民以后，一心惦记着要为耶和华的约柜建造宫殿，这是他当时最大的理想；但是在耶和华否定了他的想法之后，他没有任何的争辩和埋怨，而是完全顺从了神的旨意，在今后的日子只是默默地为宫殿的建造预备各种材料；耶和华也大大赐福给他的子嗣后代，应许由他的儿子所罗门王来完成他的未了之愿
* 2. 我们要在主面前安静谦卑下来，在主里面找到安息；这时候圣灵会在我们心中做工，主的话会向我们敞开，显明他的心意；当我们得到主的呼召，领受到主的话语，我们愿意放下自己的意志，降服在主的面前，让主的旨意得以完成

#### 2. 一个专心仰赖主的人
* 1. 我们知道，大卫有一段非常困苦的时期，就是自己的嫡长子押沙龙起兵造反，谋权篡位；大卫被流返驱逐，不得不带着亲信们逃离耶路撒冷；在逃难的路上遭受到扫落的后代示每的无情嘲讽和咒骂，他之前亲密的谋士亚希多弗也背叛了他；即便是在这么困苦的时期，大卫也是全心全意仰赖神。他从不担心战争能否取胜，因为他真切地知道耶和华会帮助自己征战，他也不怨恨自己的儿子押沙龙，反而他切切地为他的性命担忧；他也没有想着报复示每和亚希多弗，单单地把一切仰望交托给神
* 2. 我们每个人的生命都有押沙龙，紧接着还会有示每和亚希多弗，把本就糟糕透顶的情况变得更糟。但是神是我们的避难所，不论好的时候坏的时候，我们要学会放手，把一切交托给神，专心仰赖他，他必帮我们征战，成就一切在我们身上属乎他的美意。倚靠耶和华的人，好像锡安山，永不动摇

#### 3. 一个让主居心中宝座的人
* 1. 在这里我们要提到三个人：
    * 1. 示每：在大卫王逃难的时候极尽嘲讽和咒骂，在大卫王得胜归来过约旦河的时候又谄媚畏惧地来迎接；示每见风使舵，是一个凭眼见而活属乎肉眼的人
    * 2. 米非波设：约拿单的儿子，大卫爱他，他也真心诚意地认大卫为王。在大卫离开和返回耶路撒冷的路上，米非波设都想一同前往，但是他被残疾的肉体捆绑，并没有做到；就像后来软弱的门徒们在客西马尼同主耶稣一同警醒祷告，却都要睡着。心灵固然愿意，肉体却软弱了。米非波设是一个属乎肉体的人
    * 3. 巴西莱： 认定大卫为王，在大卫王逃离的时候，他不顾自身安危，供奉很多食物，也不求任何的回报；在大卫王得胜归来要善待奖励他的时候，他也别无所求，仅仅是让自己死后能葬在父母的墓旁。巴西莱是凭着信心而活的。
* 2. 我们要让主居心中的宝座，不为眼见，不为肉体，单单是凭着信心为主而活。主耶稣从来不会过分地要求我们完整完全乃至完美，因为他知道我们每个人心里面都会同时居住着示每、米非波设和巴西莱三种人格。因此我们要让主居我们内心的宝座，恳切地向主祷告，让我们内心的巴西莱增加，示每和米非波设减少

#### 4. 一个信心坚固的人
* 1. 我们要成为合乎神心意的人，在很多时候都会受到撒旦邪灵的诱惑和攻击，这个时候我们能否站立得住呢？我们的信心是否坚固呢？在这里我们要提到大卫王的三位勇士
    * 1. 约设巴设：他是“一个坐在他位子上的人”，纵然他是军队的统领，神才是军队的元帅
    * 2. 以利亚撒：在众人撤退的时候，在自己疲惫不堪的时候，以利亚撒也要紧紧握住手中的刀与非利士人征战
    * 3. 沙玛： 他站在田里为了以色列人的产业而战，并且倚靠耶和华大获全胜
* 2. 我们要学习约设巴设，把领导权交给主耶稣，用主的方式来征战。我们要在仇敌面前像树一样刚强，在主耶稣面前像虫一样柔软。每当主耶稣呼召勇士，我们愿意站出来，并且凭借主的带领站得牢靠，站得稳固；我们要学习以利亚撒，牢牢地握住那把刀，时时刻刻牢记主耶稣的话语。因为主耶稣的话语是我们的权柄，因着信心，我们将无所畏惧；我们也要学习沙玛，为了我们的产业而征战。我们的产业是主耶稣，凭着信心，没有任何的撒旦邪灵能把我们与主耶稣分开，主耶稣必与我们因信称义的人同在，直到世界的末了

以上就是钟理恩弟兄在教会营分享的信息的一个简单的总结。感谢主！
