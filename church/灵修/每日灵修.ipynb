#### `受浸重生的确据`
* 【林后5:17】若有人在基督里，他就是新造的人，旧事已过，都变成新的了。

#### `得救的确据` 
* 【约壹5:11】这见证就是 神赐给我们永生，这永生也是在他儿子里面。
* 【约壹5:12】人有了 神的儿子就有生命，没有 神的儿子就没有生命。

#### `垂听祷告的确据`
* 【约16:24】向来你们没有奉我的名求什么，如今你们求就必得着，叫你们的喜乐可以满足。

#### `得胜的确据`
* 【林前10:13】你们所遇见的试探，无非是人所能受的。 神是信实的，必不叫你们受试探过于所能受的。在受试探的时候，总要给你们开一条出路，叫你们能忍受得住。

#### `认罪的确据`
* 【约翰一书 1:9】我们若认自己的罪，神是信实的，是公义的，必要赦免我们的罪，洗净我们一切的不义

# 2. 每日读经<a name="2"></a>

丰富的恩典

09月09日 日用的饮食

经文：“我们借这爱子的血，得蒙救赎，过犯得以赦免，乃是照他丰富的恩典”(弗 1:7)。

神的恩典是丰富的，是取之不尽、用之不竭的；是体贴入微，照顾周备的。他的儿女若要与他在正当的关系中，那就要长久活在他的恩惠中。无论什么情况、什么势力都不能使我们与他的恩典隔离，或跳出他的恩典之外。因为这恩典是神用诸般智慧聪明充充足足地赏给我们的。
我们害怕信心不足吗？试探太多吗？罪过太重吗？常有软弱吗？一想到神丰富的恩典，这一切顾虑就都消除了，因为我们得救是本乎恩。主曾对保罗说："我的恩典够你用的。"今天主是照样对我们说："我的恩典够你用的。"感谢他的圣名！
一切信托他、靠赖他的人都是这应许和这事实的见证。但愿我们借着信心有更大的容量，接受他丰富的恩典。这样，我们不但不掩饰自己的罪过，反倒要承认自己的罪，他必照他丰富的恩典赦免我们。这样，我们因为信靠神，就成了蒙恩得救的人，得以承受将来极大的应许。
祈祷：主啊！你的恩典的滋味何其美；你的恩典是极其丰富的，也是充充足足地赐给我们的。我们感谢你，赞美你的名，阿们！

干渴的都当就近水

09月08日 日用的饮食

经文：“你们一切干渴的都当就近水来；没有银钱的也可以来；你们都来，买了吃；不用银钱，不用价值，也来买酒和奶”(赛 55:1)。

今天有多少人都感觉灵性上干渴饥饿，整日奔走，寻不着滋养灵性的饮食。神对这样的人说，你们都当就近我来，我有水，可以解你们的干渴；我有酒有奶，可以使你们饥饿得以饱足。没有钱的也可以来，拿你们的需要作代价，我也要买给你们，使你们饱足。
我们以往花钱买那不足为食物的东西，用我们劳碌得来的买那不能使人饱足的东西，我们真是受了人的愚弄。我们要快快回到神的里面！神说，你们要留意听我的话，就能吃那美物，得享肥甘，心中喜乐。的确，我们的干渴饥饿是因为不听神的话。因为常不听神的话，他的话就不临到我们了。正如先知阿摩司说，人饥饿非因无饼，干渴非因无水，乃因不听耶和华的话。
祈祷：主啊！今天我们到你面前来，求你对我们说话，使我们的饥渴得以消除。求你赐圣灵开启我们的心，使我们能接受你的话到我们心里，成了我们生活的力量。奉耶稣的名求，阿们！

他的话在我们的里面

09月07日 日用的饮食

经文：“你们若常在我里面，我的话也常在你们里面。你们所愿意的，祈求就给你们成就”(约 15:7)。

将神的话系在手上，戴在额上，佩在身上，写在门上，是没有多大用处的。这些作法无非是要使神的话容易进入人的心，但结果不一定如此。神的话不进入人心，是不能起作用的。手里常带着一本圣经，不如把神的几句话记到心里，并且相信奉行。
神是借着他的话与信他的人同在。神若与我们同在，我们就有权利向神祈求我们所要的，他也必为我们成全。这就是他的话在我们里面所表现的力量。
心里若不相信，神的话就只在人的口头上念诵，这样所起的作用不多。讲道的人所以不能行道，就是这个原因。
许多人的信心动摇，都是因为没有神的话在他们里面。他们所有的话，可能是牧师的话，或是书上的话，或是还没有明白的圣经上的话，但都不是神在他们心里的话。所说神的话在人心里，是听了明白的话，这样的话常常叫我们放心，不要惧怕，不要忧虑；要求神的旨意成就，并舍去自己，背着十字架来跟随他；还要爱人如己。若我们借着听见的话，希望神偏待我们，为我们行奇事，满足我们的私心，或特别抬举我们，使我们胜过别人，少有不顺，多有享乐，可以断言这决不是神借着耶稣所传的话，这都是要借着宗教信仰，成就个人的私图，是神所不喜悦的。
祈祷：主啊！求你不使我们所听见的话更叫我们自私自利。求主将你有生命有能力的话放在我们心里，使我们愿意你的旨意成就。奉耶稣的名求，阿们！

总要劳力

09月06日 日用的饮食

经文：“从前偷窃的，不要再偷。总要劳力，亲手作正经事，就可有余，分给那缺少的人”(弗 4:28)。

偷窃是卑污的行为。许多人反省自己的生平，说我一辈子没有偷过东西，这也许是实在的。然而从广义看，凡存不劳而获的心的，便涉及偷窃的行为。
人为所得的好处情愿付出代价，那是避开窃取最好的方法。各尽所能，按劳取酬，是人类社会经济生活的正当原则。一切的强取豪夺、剥削欺诈，都是从企图不劳而获的动机里出来的。劳力就是满足生活需要的正当手段。
信神的人在凡事上都不要亏欠人，所以更不能象俗语所说的“好吃懒做”；因为那是一定不能讨神的喜悦，也不能荣耀神。所以我们要劳力，要操作，要亲手作正经事，这样不但能使自己的需要得以满足，也能有余分给那些失去劳作能力、或所得不足以养生的人。世上一切勤俭人的生活都能为我们证实这个千真万确的真理。同样，你我若选择走这条康庄大道，也要达到同样满意的结果。感谢神，赞美神，使我们不但可以自给，还可以随行方便。这就是神的道、神的教、神的真理。
祈祷：求主帮助我们不再存虚妄的心行事。我们既听了你的道，领了你的教，学了你的真理，就要照着去行，亲手劳力，作正经事，不但赡养自己，也可以帮助有缺欠的人。奉基督的名求，阿们！

趁着白日

09月05日 日用的饮食

经文：“趁着白日，我们必须作那差我来者的工；黑夜将到，就没有人能作工了”(约 9:4)。

作事要成功，须趁着机会。有了机会不作，机会过去了，就追悔不及。白日就是作事的好机会，时光错过了，黑夜来临，要作也是不可能的。
我们今天作信徒的，要存着感谢的心，领受神所赐的宝贵光阴，在我们的事工上求一天的进展。如果让今天蹉跎过去，虽然明天再来，如何能保证我们好好地去利用它呢？要有决心去利用今天，对于明天就更容易把它看为作事的机会了。
基督耶稣受天父的托付，天天殷勤勉励地去完成他的使命。我们在各人的职业上，也有神特别的旨意要借着我们的事工实现出来。因此我们应当趁着机会竭力去作，好象那忠实聪明的管家时刻尽他的责任。主人来的时候看见他这样作，他就有福了：主人要交派他更大的责任。
祈祷：求主使我们顺从你的旨意，常作准备，在你所派我们的事上忠心到底，直等到主来。奉耶稣的名求，阿们！
“人出去作工，劳碌直到晚上”(诗 104:23)。
“你所作之工，必有赏赐”(耶 31:16)；
“你们都当刚强作工，因为我与你们同在”(该 2:4)。
“你们务要……，常常竭力多作主工，因为知道你们的劳苦，在主里面不是徒然的”(林前 15:58)。

他垂听穷人的祷告

09月04日 日用的饮食

经文：“他垂听穷人的祷告，并不藐视他们的祈求。……要垂听被囚之人的叹息，要释放将要死的人”(诗篇102：17-20)。

在旧社会里，人常是轻看穷人，但是神却要听他们的祷告，不因为他们的贫穷就轻视他们。穷人的需要是急迫的，他们的祈求也最诚恳的，所以神要垂听他们的祷告。
有财有势的人，朋友多，帮手也多，他们不求神，因为他们用不着神，他们能受到什么冤屈呢？谁敢欺负他们呢？他们却倚势欺人，恃强凌弱，特别是欺侮贫苦无告的人。但神却听了穷人的祷告，就在欺压人的身上替穷人报仇、伸冤。无辜被囚的人的叹息，神也听见，他要搭救他们。冤抑至死的人，神要为他们昭雪。
耶和华是赏善罚恶的神，他是主持公道、亮察严惩的主宰。他不枉法，也不徇情。他等待人悔改。但他绝不放过怙恶不悛的人。我们的心应当顾念穷人，应当代那受委屈的人喊冤，一切的不公都是人类的仇敌，一切的罪恶都是与神对立的。信神的要与神同工，促进天国的实现。
祈祷：主啊！求你垂听穷人的祈祷，顾念一切受冤屈的人，也求你帮助我们主持公道正义，以身作则，把你的公义表现出来。奉耶稣的名求，阿们！

求你坚固我的手

09月03日 日用的饮食

经文：“他们都要使我们惧怕，意思说，他们的手必软弱，以致工作不能成就。神啊！求你坚固我的手"(尼 6:9)。

要作好事，磨难就来了。尼希米要重修圣城，他的仇敌就多方阻挠。然而他所受的是神的使命，他在他所代表的民族中开始了复兴工作。这是神圣重大的责任，只能求任命他的神坚固他，领导他，克服当前的困难，完成他的使命。
别人都软弱，尼希米不能软弱；别人都惧怕，他不能惧怕。人劝他躲避到神的殿里，他说：“象我这样的人，岂要逃跑呢？岂能进入殿里保全生命呢？我不进去。"
感谢神的大恩，凡求告他的，他就加强他们的信念，壮起他们的胆量。他们才能修好了城墙，订立了新约，除掉了对于贫民的剥削，割断了与外族联姻的关系。
今天信神的人也当求主施恩，坚固我们的手，使我们在建设教会及建设国家的事工上积极努力，多有贡献。也使人人守诚实、行公义，离开一切邪恶，谨守遵行神的诫命，长久蒙福，住在神赐给我们的地上。
祈祷：主啊！我们愿意象你的仆人摩西那样地祷告说：愿主我们神的荣美来到我们身上。愿你坚立我们手所作的工；我们手所作的工愿你坚立。阿们！

必蒙拯救

09月02日 日用的饮食

经文：“行动正直的，必蒙拯救；行事弯曲的，立时跌倒。……心中自是的，便是愚昧人；凭智慧行事的，必蒙拯救”(箴言28：18，26)。

不受磨难，不能发达，这是人生的原则。然而能得着磨难的益处的却是行善的人，作恶的人常被磨难所销毁。
神若常与我们同在，那些磨难便能充实我们的优点，增加我们的用处。这就是神的最有效力、最觉特殊的拯救。神拯救他的先知、使徒和独生爱子，就是用这个方式。神能拯救信他的人脱离疾病、灾难、危险；然而这些还是拯救的初步。他要救我们到底，就是使我们成圣，完全合乎他的用处。
善人跌倒，还要起来；恶人陷溺，不能自拔。善人行善，处处求生；恶人为恶，自速其死。这是天理的循环，也正是神的报应。敬畏神是智慧的开端；远离恶事便是聪明；自以为聪明，偏行不义的事，那就是愚昧人。他们居心是害别人，结果却害了自己。箴言二十八章说：“人以厚利加增财物，是给那怜悯穷人者积蓄的。”这是剥削者所没有料到的。
祈祷：求主警戒我们不走弯曲的路，不作愚昧的事，不损人利己；也使我们相信，你在我们遇患难的时候必然拯救我们。奉主的名求，阿们！

他以慈爱救赎我们

09月01日 日用的饮食

经文：“他们在一切苦难中，他也同受苦难，并且他面前的使者拯救他们；他以慈爱和怜悯救赎他们”(赛 63:9)。

神用慈爱怜悯把他的百姓从为奴之地救出来，在他们面前行了许多神迹奇事，并且多方供给他们，保护他们，面陆走海，战胜敌人，得了美地为业。
今天我们信的人也是同样受了这样的恩惠，应当从心里感激神。他还有更大的慈爱要施展在我们身上。我们若准备接受他的恩爱，他就用美物使我们的心愿满足。
以色列人虽然蒙了神的慈爱，却未能长久地得到神恩上加恩，反倒因了他们的恶行，使神的慈爱和怜悯中断了，这真是可惜！他用慈爱救赎他们，本是要使他们成为大国，永远事奉独一真神，引导全世界的人归向他；但是他们却未能成就神的旨意。
我们今天受神恩惠的人也当反省，纠正我们的错误，承认我们的罪，悔改我们的恶行，立一个新志愿，努力实现他爱的旨意，在世上为他的诚实、公义、圣洁、慈爱和他的救赎作荣耀的见证。
祈祷：感谢我们的天父，你用了你的慈爱救赎了我们，但愿你引导我们成就你拣选我们、爱护我们的美意，使我们在世的日子，为你的慈爱和怜悯作见证。奉耶稣的名求，阿们！

必靠他欢喜

08月31日 日用的饮食

经文：“我们的心必靠他欢喜，因为我们向来倚靠他的圣名”(诗 33:21)。

写诗的人在这篇诗篇上说：“靠马得救是枉然的，马也不能因力大救人。耶和华的眼目看顾敬畏他的人和仰望他慈爱的人，要救他们的命脱离死亡，并使他们在饥荒中存活。我们的心向来等候耶和华，他是我们的帮助，我们的盾牌。”我们知道君王要靠兵多得胜，勇士要靠力大得救，而信耶和华的人只能靠耶和华得救。我们的救恩是从神而来的。惟独神能施行拯救，神若拯救我们，我们就真的得救了。别的能力救人都有限制，惟独神救人没有限制，是救到底的。他能借着世人的力量救我们，他也能用他自己的能力救我们。
有时候他救我们的方法不是叫我们脱离患难，乃是叫我们胜过罪恶。这样，不但救了我们自己，也借此救了许多人。所以，我们的心靠他欢喜，随时随地都可得平安。信他的人他必不丢弃，他是无所不能的神，有充分的智慧和能力，保护信他名的人不受损害。
祈祷：求主救我们脱离一切患难，使我们信你的心一天比一天加强。我们知道你是可倚靠的全能的神，求主使我们因你而欢喜。奉耶稣的名求，阿们！

保守我在他的亭子里

08月30日 日用的饮食

经文：“因为我遭遇患难，他必暗暗地保守我；在他亭子里，把我藏在他帐幕的隐密处，将我高举在磐石上”(诗 27:5)。

患难来了，人要躲避。信靠神的人要躲藏在惟一可靠的神那里。这段经文更美妙地说：神要把我们藏在他的居所，不是我们跑了去，好象是他把我们带了去。
隐藏在神的居所，真是平安幸福之至！因为在那里不但全能的神作我们的保障，而且我们还能常常看见天父的面。感谢赞美他的全能与大爱。
他保守我们在稳妥的地方，使祸患不临到我们。在那里我们也不寂寞孤独，因为我们的天父与我们同在。天父离我们越近，危险离我们越远。我们在神保守的地方还可以欢然献祭，唱诗歌颂救我们的主。
我们遭遇患难的时候，他就来搭救我们。他来的时刻是最相宜、最美妙的时刻。他的方法也是最有效、最能造就人的方法。他的眼不盲、耳不聋、臂不短，他必施行拯救。
祈祷：求主在患难临到我们的时候，把我们藏在你的居所，使我们得着你的造就，不致受患难的损害。奉耶稣的名求，阿们！

在苦难中我拣选你

08月29日 日用的饮食

经文：“我熬炼你，却不象熬炼银子。你在苦难的炉中，我拣选你”(赛 48:10)。

神有事工要成就，所以他要拣选人作他的工。他所拣选的人，因为含有不纯洁的品质，不合大用，所以他要熬炼他们，使他们精纯，能担负重大的责任。

#### 如果我的恩赐不止这一个就好了，拥有多些不是更好吗？

12:6-8 看看这里所列出的，想一想什么人可以拥有这些恩赐。先知要有胆量，勇于发言；执事需要忠诚；教师要有清晰的头脑；劝导的要知道如何打动人心；捐助者需要慷慨有信心；领袖要有行政的头脑；安慰者则在别人有需要的时候愿意关怀他们。上述所提到的所有恩赐，不可能由一个人完全拥有。一个坚持己见的先知不一定是个好的劝戒者，慷慨的捐助者也可能不是称职的管理人员。你如果已经知道自己的恩赐，就要思想该怎么样使用这些恩赐来兴旺教会，与此同时还要时刻提醒自己，个人的恩赐不可能独力承担整个教会的工作。应当为其他人拥有跟你不同的恩赐而感恩，以你之长补他之短，并且甘心情愿地让他之长来补己之短，这样你们就能一起建造教会了。

#### 有些人妄自菲薄，有些人自视过高，如何可以不卑不亢？在基督里的人是……

12:3 正确地估计自己是极其重要的，因为有些人过于轻视自己，有些则过高地估计自己。一个准确的自我评价必须基于对自我价值的认识，这是指在基督里的身分。离开了基督，在永恒之中我们就所值无几；但是在基督里，我们就价值连城，而且有事奉的能力。倘若你只从世界的成功标准来评价自己，就看不到你在神眼中的真正价值了。

#### 不要效法这个世界，就不做其上的事；至于心意更新……

12:2 这个世界的行为与习俗，绝大部分是自私且腐败的，所以聪明的基督徒都不应该效法。但是我们拒绝和世界妥协，却不单是局限在行为和习俗上，更重要的是在思想方面。所以，这一节也可以翻译为“不要向世界妥协，要更新改变你的思想”。一个脱离了世俗行为的人，仍然可以是一个自私、贪婪、骄傲、顽固和狂妄的人。惟有我们的思想被圣灵更新、改造和重新建立之后，我们才得着真正的更新（8:5）。

无边无际的阳光，照耀着广袤寂寥的原野。庞大的收割机一路翻卷着麦穗，在金色的麦田里碾出笔直清晰的痕迹。混合着吉他与口琴的背景乐重复着简洁质朴的音符，淡淡的，懒懒的，像极了天边舒卷的云彩。路的远处，发须花白的史崔特先生驾着那辆66年产的绿色铲草机，缓缓，缓缓的驶入我们的视野…… 《史崔特先生的故事》是一部典型的公路片，但它又有着明显区别于同类型电影的非典型因素：踏上漫长旅途的并不是某个血气方刚、迷茫焦躁的年轻人，而是一位年逾古稀、久病缠身的老人家；交通工具也不再是帅气十足的敞篷轿车或越野吉普，而是一台破旧二手铲草机所改造的小小拖车。 横看竖看，史崔特都不是一个适合独自旅行的人，尽管他身材高大，穿上牛仔裤和夹克衫，颇有些牛仔范儿，但73岁高龄、糖尿病、视力衰退、早期肺气肿、严重盆骨问题等生理上的病痛早已压得他喘不过气，而路途遥远、没有驾照、经济拮据等客观困难又横亘在他的面前。是什么让他下定如此大的决心，去完成这样一段艰难得几乎绝望的旅程？如果只是为了探望中风的哥哥，他理应选择一种更加快捷的交通工具，可他偏偏骑上了一辆和他同样衰老的铲草机，速度慢得连自行车都能从一旁呼啸而过。为什么？我想，这就像问唐僧为何不让孙悟空一个跟头翻到天竺取了经书了事，而非要一步一挪，历经九九八十一难，才肯领受那普度众生的大乘佛法。有时候，地理上的目的仅仅是旅行的一个方向，而行者真正要达到的终点，其实就是他们内心的信仰。漫漫旅途中的期待、求索、挫折、思考以及成长，皆是他们迈向信仰之门的斑驳足迹。  也许在旅途的起点，就连史崔特自己也并不清楚自己的想法，也许那时在他心中涌动的仍是一股不服老的倔劲。可是，当他终于上路，当他独自穿过一片片辽阔的田野，送走一个又一个清晨与傍晚，他的心也渐渐擦亮。劝说离家出走的少女，他显出一位父亲对子女的慈爱；点拨争吵的兄弟，折射出他对哥哥的内疚与歉意；与同龄者回顾峥嵘岁月，他的眼角闪现愧对战友的悲痛泪光……每一个人都像一面镜子一样反映出史崔特的某个侧面，他也在这些投影中找到真实的自我，咀嚼出人生的况味。最令我动容的，是他与一群自行车赛手的对话，充满青春气息的少年人一边玩闹，一边无所顾及的抛给他一个近乎残忍的问题：“变老之后，最糟糕的事情是什么？”史崔特略一停顿，低沉的答道：“最糟糕的，便是想起自己曾经年轻的时候。”此时，他苍老的脸在火光中若隐若现，显出一种浸人的悲凉。 好在，哀伤并不是人生的全部，年龄的增长虽然伴随着体能的衰退，但也收获了丰富的阅历与感悟，就像深秋，在为大地带去冬日寒意的同时，也给世人送来丰收的果实。途中，史崔特曾遭遇了一个奇特的小插曲，一位妇女驾车撞死了一头鹿，她对着前来帮忙的史崔特歇斯底里的抱怨一番之后，又绝尘而去。无奈的史崔特默默看了看死去的鹿，当晚便将晚饭的菜单改成了烤鹿肉。这看似一个幽默片断，却体现出一种只有老人才能拥有的对生命的豁达态度，比起任由这头无辜的鹿在路边腐烂，遭来往车辆肆意碾压，不如为它赋予一些实用的价值，也算死得其所吧。还有一次，乌云压境，暴风雨转眼即至，史崔特赶紧将拖车驶入路边一个废弃的仓库。待到瓢泼大雨从天而降，他便点燃一只雪茄，坐在自己的“坐骑”之上，悠然欣赏起他最喜爱的电闪雷鸣。这样的泰然自若完全不输于牛仔驰骋原野时的冲天豪气，叫人看了更加喜欢上这个可爱的老头。 历经6个星期，史崔特终于完成了长达370多公里的旅程，来到哥哥莱尔所居住的木屋旁。兄弟俩十年来第一次见面，一个拄着两条拐杖，一个撑着助行器，老眼相对，泪光隐现，早已无需言语。最后，镜头从史崔特先生的近景缓缓上摇，直至叠印出那片满是星斗的深色苍穹，影片也在这一片静谧中悄然落幕。同样上升的，还有史崔特内心的信仰，那是对亲情、自我以及人生的充分肯定。 不同于大卫·林奇其他作品的诡异色彩，《史崔特先生的故事》显得朴实沉稳，大量的叠印不仅完美展现了乡间优美的风景，也将老史崔特内心的伤感、困惑、豁达和淡定充分的展现了出来，内容与形式得到完美的契合。而安杰洛·贝德拉曼堤那充满空气特质的优美配乐，更是为影片平添了不少温馨的感动。在绝大多数公路片中，主人翁的上路都是源于一种青春的冲动和困惑，他们的不羁驰骋是对于庸常人生的反抗和嘲笑，对于不公命运的愤怒与诘问。而在本片中，史崔特的上路则像是一种静修，在漫长的旅途中徐徐而行，不仅将途中风物尽收眼底，更要细细评出其中真味。对生命意义的追寻，本来就没有年龄的限制，只要心中有信仰，便可随时上路。

#### 18:1 “常常祷告，不可灰心” （路加福音）

意思不是说我们要说重重复复的话或是举行冗长无聊的聚会，而是说我们每日为神而活时，就要不断将我们的要求带到神面前，相信祂会答应我们。神会延迟答应我们，但祂的延迟常常都有更美好的旨意，我们不可认为是祂疏忽了。当我们恒久祷告时，我们的品格、信心和盼望都会成长。

#### 路加福音 9:5 

耶稣让他们跺掉脚上的尘土，具有深远的意义。这行动表明门徒不用对这些人如何回应他们的信息负责。只要我们忠心地传扬基督，即使被人拒绝，我们也用不着负责。像门徒一样，我们只须继续前行。

#### 神的管教：
关于神的管教的问题， 也许我们不容易引若干圣经节。现在，我们先看一段圣经，就是希伯来书十二章四节到十三节，然后再来看管教的意义到底是如何。

##### 1.管教的意义

第四节：“你们与罪恶相争， 还没有抵挡到流血的地步。”在这里，你看到使徒对希伯来人说，你们与罪恶相争，虽然受了很多的苦，受了很大的试炼，也碰着了各种的难处，也遭遇了许多的逼迫，但总还没有抵挡到流血的地步。这一个，如果和我们的主来比较的话，是差的太多了！本章第三节告诉我们说，主耶稣是轻看羞辱，忍受了十字架的苦难，到了一个流血的地步。信徒们，虽然也受了羞辱，也忍受了十字架的苦难，但是还没有到流血的地步。

* 为什么要碰着难处

一个人作了基督徒，他所要预备遇见的是什么事？我们绝不愿意把一个不正当的盼望，摆在弟兄们面前；乃是要给他们看见，他们将来要碰着许多事，那些事都有神的目的在里面，都有神的意义在里面。他们要遇见许多试炼，他们要遇见许多难处，但那些试炼，那些难处，到底为这什么缘故？到底有什么意义在里面？要给他们看见基督徒的苦，那一个苦的意义在那里？当主还没有留我们作殉道者之前，无论如何，抵挡罪恶，与罪恶相争，还没有到流血的地步。我们所遭遇的羞辱，我们所遭遇的十字架的苦难，就是与罪恶相争，抵挡罪恶，虽然没有到流血的地步，但是，是在那里抵挡！
到底这一件事的发生，是什么缘故呢？
五至六节：“你们又忘了那劝你们如同劝儿子的话，说：我儿！不可轻看主的管教，被他责备的时候，也不可灰心；因为主所爱的，他必管教，又鞭打凡所收纳的儿子。”

* 不能盼望走顺路

使徒在写圣经的时候，引旧约箴言的话对我们说，主如果管教我们，我们不可轻看，主如果责备我们，也不可灰心。这两个态度，都是信徒该有的。相反的，都是有人经过他的困难的时候，当他在那里受苦的时候，有神的管教在那里的时候，它看作平常，以为不是重要的事，很轻易的就让神的管教从他身上过去。或者有人当他碰着主的责备，有主的手加在他身上的时候，他灰心了！他觉得说，一个基督徒的环境这样困难，作基督徒是太难了。他还盼望说，这一条路是非常顺的，是穿了洁白的细麻衣，很轻松地走在黄金的街道上，走到珍珠的门里头去。他没有想到作基督徒的时候，要碰着许多为难的事。他没有预备好要在这一种环境里来做基督徒，他觉得灰心，他觉得这条路不大好走。箴言给我们看见这两个态度都是错的。

* 管教的目的

人不应该轻看主的管教。主如果管教我们，我们要认真。主叫我们所遭遇的事，都有它的目的，都有它的意义在里面。他要借着这些遭遇，这些环境，来造就我们！他所有的管教，都是为着成全我们，都是为着叫我们圣洁。他所有的管教要叫我们在性情上有了神的性情，在品格上受了神的教育。这些是主管教的目的。主不是无缘无故地打我们，主是要把我们打成功作一个器皿。主不是没有意义地叫他的儿女受许多苦。光受苦不是受苦的目的，不是光要你受苦，所以给你苦。受苦乃是有目的的。受苦是要叫我们与神的圣洁有份，受苦是要叫我们与神的性情有份，这一个是管教的目的

* 要重看神的管教

可是有许多神的儿女，很可能作基督徒作了十年八年，还没有一次重看神的管教，还没有一次在那里说，主是在这里管教我，主是在这里对付我，主是要把我对付到一个地步，主是要把我打到一个地步，打成功一个器皿，雕刻成功一个器皿。他没有看见神的打，神的对付，神的雕刻的目的，他就是这样马马虎虎的过去。今天碰着一件事，他就是无所谓地让它过去；明天再碰见一件事，他也不知道主的目的到底是在哪里，也就随便让它过去。好像说，神是一位没有目的的神，是随便给人苦受的。所以，请你们记得：第一个态度，就是神的儿女应该学习重看神的管教。无论那一天一有遭遇，第一个问题要问：“这是什么意思？为什么要这样作？”我么要学习重看，不要轻看。轻看就是马虎，轻看就是说神要怎么作，就怎么作，自己莫名其妙的，迷迷糊糊过去就是了。

* 学习接受管教

一面，轻看是不应该的。但另一面，也不要把管教看得太重。作了基督徒，一天到晚，就是受苦，就是不顺利，因此就灰心了！这是把管教看得太重了。我们要学习接受主的管教，要学习看见主的管教是有意义的。我们不要灰心，要学习接受。

##### 2.管教的性质
第六节：“因为主所爱的，他必管教，又鞭打凡所收纳的儿子。”这是箴言的话。这给我们看见说，所有的管教，到底是为什么缘故。

# 4. 从以撒和路得的故事，看圣经中找配偶的原则<a name="4"></a>

从以撒和路得的故事，看圣经中找配偶的原则
文/周晓勤

生命季刊微信专稿

 当今社会物欲横流，性爱泛滥。找配偶，以物质为筹码，以感觉为向导，已经到了恶脓要穿孔爆流的地步。很多人一边怨声载道叫苦连天，一边却添柴加油推波助澜。作为基督徒，一群被上帝从世界选招出来，身居世界，心却不该属世界的人，在这恶俗的洪流冲击下，怎样才能不同流合污，随波逐流？圣经在找寻配偶的事上，是如何教导的呢？

圣经上讲了两个经典的找配偶的故事，一个是男方主动提出，就是以撒；一个是女方主动提出，就是路得。这里以以撒和路得的故事为例，看看圣经上关于基督徒找配偶的原则。首先要声明，因为是叙事体裁，有些故事情节被当时的文化及历史背景笼罩着，乍一看，不太明朗。但我们可以把其中的原则抽剥出来，在现今的社会里应用。

（以撒和路得的故事，请读创世记24章和路得记全书。基督徒只能找基督徒作配偶，信与不信不能同负一轭的原则，请读林后6：14-18；林前8：39；创224：3-4；27：46；申7：1-4；22：10；书23：11-13；士3：4-7；代下18：1；19：1-2；21：1-6；拉9：11-14；尼13：23-27；诗106：34-36；结44：22-23。）

 

1.基督徒只能找基督徒作配偶，信与不信不能同负一轭。

 

亚伯拉罕严肃认真，发誓赌咒地要求他信任的老仆人，不畏艰难，千里迢迢去他的老家美索不达米亚（今伊拉克境内），在亲戚的女儿中，为爱子以撒找一个妻子。要知道四千年前，没有飞机火车汽车，在中东的沙漠地带，长途旅行全靠骆驼，不是一件容易事。他绝不就地取材，在他生活的迦南地（今以色列境内）为儿子娶妻。要知道亚伯拉罕在迦南地是德高望重威震四方的大富豪，他曾带领自己的家兵家将，一次就打败了迦南地的几个王。只要他就地开门纳媳，保证当地名门旺族的漂亮女儿们，被父母或媒人领着排队来相亲。因为以撒是他产业的唯一继承人，是不折不扣的富二代贵二代（创24：2-4；14：13-24；23：6）。亚伯拉罕不就地取材的原因很简单，因为迦南地的女子不信靠上帝。只有到自己的老家亲戚中，才有可能找到信靠上帝的女子。亚伯拉罕的决心之大，令人敬佩，他情愿儿子一时半会不结婚，成为大龄青年，也不妥协迁就，坚决不娶不信靠上帝的女子。

 

路得也是持守同样的原则。她的丈夫死后，她年轻且无孩子拖累，她本来可以选择听她婆婆的劝告，像她妯娌一样，回到不认识上帝的摩押娘家，重新嫁人生孩子。当然，那样嫁的丈夫肯定是不认识上帝的。但她立定心志跟着婆婆回犹大地，追随嫁到婆家后才认识的造物主上帝。也就是说，如果她以后有幸再嫁，就可以嫁一个认识上帝的丈夫。

 

2.人要先“离开”父母，有独立的人格，再进入婚姻。

 

圣经讲婚姻的真理言简意赅：“人要离开父母，与妻子联合，两人成为一体”。就是说，人在进入婚姻之前，必须在情感理智及经济上与父母分开，成为独立的个体，然后再同配偶联合。中国文化因为不懂上帝设计婚姻的真道，强调“血浓于水”，把亲子关系放在夫妻关系之前。往往身体上已经成年的孩子，心理和经济上离不开父母。很多父母（特别是母亲）也拒绝离开孩子，不愿意接受孩子婚后与自己的关系有别于婚前。结果就是婆媳之间张力很大。所谓的婆媳矛盾，只是表象，其根源是婆婆不经意地充当了儿子媳妇婚姻中的第三者。

 

以撒出生时，他爸爸亚伯拉罕一百岁，妈妈撒拉九十岁。而且他是他的富豪爸爸妈妈盼了一生才盼到的唯一儿子，在父母和众家仆眼里，他千万恩宠集一身，说不尽有多宝贝。亚伯拉罕在摩利亚山上通过上帝的考试后，不再把以撒当命根子了。他能够在心理上放下以撒，让他在上帝面前成长为独立的个体。但可以猜测妈妈撒拉很娇宠依恋以撒，以撒在心理上也离不开母亲。因为撒拉过世时，以撒已经37岁，尚未考虑结婚，要知道当时37岁的男子早做爸爸了（创11：12-24）。而且母亲死了三年后，以撒40岁结婚时，才从丧母的悲痛中走出来（创24：67）。可见以撒当时仍缺少独立生活的能力，尚未成熟到进入婚姻的地步。三年后，亚伯拉罕才开始帮他张罗婚姻。

 

为人父母，如果真正的爱自己的孩子，就要早早地有计划有步骤地训练孩子独立。因为上帝命令“人要离开父母”。而不该拽着孩子不放手，以满足自己心理情感上的私欲。父母的这种私欲，往往被孩子误认为是爱，很难拒绝，很难识别。这样危害孩子的婚姻，还让孩子理不清头绪，不知道矛盾的根源在哪儿。

 

路得就不用说了，因为她很早就离开父母了。

 

3.在正确的地方找。

 

只有在正确的地方找，才有机会看清品行；只有在正确的地方找，才能找到愿意在正确的事上花时间和精力的人。老仆人在井边为以撒找配偶，因为那时候在沙漠地带，井是人和牲口的主要水源，是人和牲口聚集的公共场所。是女孩子们干活的地方，也是她们聚集聊天的地方。在那儿可以观察到一个女孩子是否勤快，是否乐于助人，以及她如何跟其他人打交道等等。

 

路得和波阿斯是在收割庄稼的田里相识。在不短的收割季节里，波阿斯看清路得是一个依靠上帝，殷勤努力，吃苦耐劳，规矩正派，爱婆婆如己的好女人。路得也有机会看清波阿斯是一个遵循上帝旨意，爱人如己的人。

 

现代生活哪些地方是找配偶的正确地方？应该是一个公共场合，一个可以观察到对方做人做事的态度的地方，一个可以看见对方跟他人打交道的地方。比如教会，做义工的场合，学校，工作场合，大家庭聚会等等，不应该是酒吧，舞厅，赌场，风月场所等等。

 

多年前我认识一家美国人，妻子年轻时曾是拉拉队队员，40多岁了还亭亭玉立，妩媚动人。夫妻俩是在酒吧相识结婚的，因为两人都喜欢在酒吧喝酒跟陌生人聊天打发时间。两人有一个女儿，丈夫工作需要经常出差。每次丈夫出门在外，妻子就把4岁的独生女儿借宿在保姆家，独自去酒吧喝酒聊天打发时间，因为她耐不住寂寞，这样迟早会红杏出墙。记住，你用什么东西把对方吸引来的，你就得继续用同样的东西维系，有时甚至要加码。

 

4.看重内在的品格，而非外在的条件。

 

如果你是一个仆人，奉王命为王子物色一位王妃，你会找什么样的？几乎所有的人都会答：

 

找最美貌的！

 

我真的佩服亚伯拉罕培养熏陶的这个仆人的属灵成熟度，看看他祷告求什么样的女子：“耶和华我主人亚伯拉罕的神啊，求你施恩给我主人亚伯拉罕，使我今日遇见好机会。我现今站在井旁，城内居民的女子们正出来打水。我向哪一个女子说：‘请你拿下水瓶来，给我水喝’，她若说：‘请喝！我也给你的骆驼喝’，愿那女子就作你所预定给你仆人以撒的妻。这样，我便知道你施恩给我主人了。”

 

他祈求的这个女子，向她要一点水给人喝，她不但乐意给，还主动提出从井里打水，喂饱十匹骆驼！一个人喝几杯水就够了，一匹骆驼可要喝几桶水，何况是十匹骆驼！要花多长的时间，这个女子要有多大的爱心和耐心！是多么的吃苦耐劳，勤劳肯干！而且自己眼里有活儿，不需要别人安排，更不需要别人跟着背后催。她能一口气喂饱十匹骆驼，身体应该也不赖。这种女子太适合做一个大家业的女主人了！他求品格，上帝给了，还顺便给了美貌为额外奖励，利百加“容貌极其俊美”（创24：16）。当然，给不给额外奖励完全是上帝的主权，不在我们的掌握之中，但基督徒在找配偶时，应该注重正确的东西。

 

路得的经历也是如此。路得在波阿斯的田里拾了一段时间的麦穗，通过自己的观察和婆婆的分析，都认定波阿斯是一个敬畏上帝，活出信仰，爱人如己的男人，虽然波阿斯的年纪有点太大（路得记3：10）。

 

当然，如果你想找一个敬虔爱上帝，品格高尚的基督徒配偶，自己必须先做一个这样的基督徒。因为什么饵，就钓什么样的鱼；什么鞍，就配什么马。如果你现在尚未发现任何人，这是上帝给你的时间和机会，不要在自怨自怜中浪费了。要一边殷勤地为那尚未出现的另一半祷告，一边好好地在与主的关系以及个人品格上准备自己。

 

5.请成熟的基督徒帮助参谋把关。

 

人很容易随着自己的眼目和心意乱跑（参见民15：39），在男女关系上尤其如此。一喜欢上对方，就坠入爱河。关闭了理智的眼睛，打开了荷尔蒙的闸门。因此，很难看清对方的品格缺陷，以及彼此是否合适做夫妻。以撒有敬虔爱上帝的爸爸亚伯拉罕和忠心智慧的老仆人做参谋，路得有明白上帝心意又爱自己的婆婆做参谋。

 

我做年轻人的辅导时，建议他们如果觉得自己对某弟兄或姊妹有些意思，不光要祷告问上帝，也要找教会或团契里了解自己、也了解对方、比较成熟的基督徒咨询一下。问正确的问题：你觉得我们俩彼此是否合适以婚姻为目的的交往？你觉得我们俩的属灵光景是否般配？你觉得我们俩彼此的长处和短处是什么？能不能请你去帮我问问他对配偶的期望？等等......。千万不要等到谈婚论嫁时才咨询，因为那时很可能已经太晚。

 

6.有父母的认同和祝福。

 

很多人年轻无知时认为婚姻只是两个人的事，结婚后才发现，许多时候吵架的导火线是因为对对方的家人有苦毒。婚姻必须要有父母的认同和祝福，这样以后的婚姻中就会少一些定时炸弹，多一些祝福和帮助。并且，基督徒年轻人因此才能在非基督徒父母面前做美好的见证。以撒有父亲的祝福，利百加有父母及哥哥的祝福，路得有婆婆的祝福。

 

如果碰上故意刁难的父母怎么办？基督徒年轻人要祷告求智慧，用温柔的心劝导父母。如果父母坚持自己的意见，而自己却认定对方就是上帝安排的配偶，怎么办？既然相信是上帝安排的，就要相信谁也不可能拆散的，包括父母。父母现在持反对意见，也在上帝的掌控之下。有两种可能：一是当事人自以为是上帝配的，其实并非如此，上帝借父母拆开关系；二是的确是上帝配的，但上帝借父母的刁难，磨练塑造基督徒忍耐等候的品格。遇到父母的坚决反对，双方应该暂时停止交往，在上帝面前祷告安静等候。如果不是上帝的心意，过一段时间，冷静下来后关系自然冷淡下去。如果是上帝的心意，就继续等。边等待边恳切祷告，求上帝改变父母，也求上帝让自己看见自己有什么需要改变的。等待期间不要同别人交往，一直等到父母同意。要相信上帝会在父母身上动工的。李秀全林静芝牧师师母夫妇在这方面做了美好的见证，他们恋爱马拉松8年后，李牧师的父母才同意。

 

7.相信并跟随上帝的带领，结果也交给神，不要患得患失。

 

 

如果我按照上帝的原则相中一个，但人家不同意，咋办？亚伯拉罕说：“倘若女子不肯跟你来，我使你起的誓就与你无干了，只是不可带我的儿子回那里去。”（创24：8）我们只需要在上帝面前按真理做自己该做的那部分，上帝掌管结果。如果患得患失，就等于不信任上帝。亚伯拉罕在上帝面前的认定是，我积极主动地遵行上帝的旨意，做我的那部分。即使儿子一时半会找不着配偶，也不能搬回老家。因为搬回老家跟上帝在他们家族上的旨意相违背，得罪上帝。

 

一百多年前，内地会的创始人英国人戴德生，在他的成长过程中，一直有从上帝来的非常清晰的呼召，就是要去中国传福音。但他当时深爱着的女朋友没有这个呼召，不愿同去当时贫瘠落后的中国。戴得生只好忍痛割爱，义无反顾地去了中国,后来上帝在中国为他预报了一位有同样心志的女宣教士妻子。

 

路得的故事也是如此。路得听从了婆婆合上帝心意的指导后，就平心静气地等候上帝的作为（路得记3：18）。

 

8.男女角色有别。

 

“利百加举目看见以撒，就急忙下了骆驼，问那仆人说，这田间走来迎接我们的是谁？仆人说，是我的主人。利百加就拿帕子蒙上脸。……以撒便领利百加进了他母亲撒拉的帐棚，娶了她为妻，并且爱她。”（创世纪24：64，65，67）。利百加千里迢迢来的目的，就是很快要嫁给以撒。为什么她却多此一举，看见以撒，马上“拿帕子蒙上脸”？

 

因为在婚姻中，上帝分配给夫妻双方的角色不同。丈夫要爱妻子，是主动和领袖的角色。妻子要顺服丈夫，是响应和辅助的角色（弗5：22-33）。因此，一般人都可以观察到，不管夫妻双方是否认识上帝，脾气性格如何，能力差异如何，在婚姻中，女人最渴慕的是被丈夫呵护，男人最期望的是被妻子尊重敬佩。一个有属天智慧的女人，不论自己多能干多自立，要常常往后退几步，给丈夫空间成长，给丈夫机会做领袖。这意味着很多时候，要给丈夫空间犯低级的错误，以温柔的心辅助丈夫。祷告求上帝帮助，管住自己嘴巴，不要拆毁丈夫，要建造他。

 

我见过不少阴盛阳衰的夫妻，往往都是男人个性比较懦弱黏糊，女人能干利索。这种女人往往缺乏耐心等待丈夫成长成熟，自己冲锋陷阵到丈夫前面。她们的理由往往是丈夫不行，我不做谁做？往往这种妻子心里的苦，只有自己知道。不按照上帝的定旨生活，一定吃苦果。因为没有人能超越上帝对人的设计，不管你是否认识造你的上帝。

 

路得也是如此，她按照上帝的旨意向波阿斯求婚后（得3：9；结16：8），就不再吭声。静静地等候波阿斯处理关于产业，法律，继承权等等的事宜，而且她根本不知道自己的求婚是否成功。

 

9.婚前保持圣洁

 

自己读这两个故事，就一目了然。圣经说只要不在婚姻内的性行为都是淫乱都得罪上帝（来12：14；13：4）。这里我想谈一下在性泛滥的洪流下，如何才能保持婚前圣洁。现代社会，除了灵性的败坏以外，确实存在一些具体的诱惑。如果事先了解预防的话，可以逃避诱惑，避免很多陷阱（林前6：18）。一个诱惑就是到处都是性方面的眼目刺激，另一个就是两人有太多单独相处的机会。我一般给年轻人的建议是，两个人尽量在公共场合谈情说爱。这样一举两得，既逃避诱惑，又能比较全面地观察对方。不要去刺激感观的地方，比方说一起看比较刺激的电影录像等等。第三个诱惑就是谈恋爱的时间太长，或者说成熟度还不够，就过早进入恋爱关系。

 

我建议双方一有意向要往婚姻方向发展，就应该找成熟的基督徒做以圣经为原则的婚前辅导。不要等到要结婚时才接受辅导，很可能已经太晚。我的一个年轻朋友，认识对方不到两个月，两人就开始接受婚前辅导。认识不到一年就要结婚了，因为婚前辅导帮助把该了解的领域尽快了解。不合适的话，就尽早终止关系。如果双方没有意向向婚姻方向发展，就应当马上停止关系。我赞成我的一个同学说的一句话：“凡不以结婚为目的的谈恋爱，都是耍流氓！”

 

何况，如果你婚前不愿意约束自己的身体的话，结婚后，你的配偶怎能信任你？！出门在外，生病，或妻子生孩子时，你怎么愿意守，或怎么可能守得住？！往往有过性经验的人，比没有的人，持守起来更具挑战性。
