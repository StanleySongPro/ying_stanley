```python
df_tmpt = pd.read_csv('cash_event_20190430interest(out).csv', skiprows = 3)
df_tmpt
```

```python
df_output = pd.DataFrame(columns = df_tmpt.columns)
# df_output
```

```python
# in_principal

def in_principal(inputStr):
    if 'fam go fund' in inputStr.lower():
        return "FGO"
    elif 'fam go plus fund' in inputStr.lower():
        return "FGOPLUS"

df_output['#in_principal'] = df_DC['Account Name'].apply(in_principal)
# df_output
```

```python
# in_book

def in_book(inputStr):
    m = re.search(r'(?P<currency>\w+)$', inputStr)
    currency = m.groupdict()['currency']
    if 's&r' in inputStr.lower():
        return 'SNR_{}'.format(currency)
    else:
        return 'BK'
```

```python
# df_output['in_book'] = list(map(lambda x,y: x+'_'+y, 
#                                 df_output['#in_principal'], 
#                                 df_DC['Account Name'].apply(in_book)))

df_output['in_book'] = df_output['#in_principal']+'_'+df_DC['Account Name'].apply(in_book)
# df_output
```

```python
# in_strategy

df_output['in_strategy'] = 'NONE'

# in_prime_broker

df_output['in_prime_broker'] = 'SCB'

# in_pb_account

df_output['in_pb_account'] = df_output['#in_principal'] + '_' + df_output['in_prime_broker']+'_ACCT_ID'

# in_cpty
df_output['in_cpty'] = df_output['in_prime_broker']

# in_amt
df_output['in_amt'] =  df_DC['Payment Amount']
```

```python
# in_event_type

Event_type={
    'SSG0340': 'CASH_TFR',
    'CASH FEES': 'CASH_FEES_BANK',
    'CFINV': 'CUST_FEE_MAN',
    'FSINV': 'FUND_SVCES_FEE',
    'MGMTFEE': 'MAN_FEE_MAN',
    'TRUSTEE': 'TRUSTEE_FEE',
    'CREDIT INTEREST': 'INT'
}

def event_type_map(inputStr, Event_type = Event_type):
    try:
        return [v for k, v in Event_type.items() if k.lower() in inputStr.lower()][0]
    except:
        return np.nan

def in_event_type(df_output, df_DC):
    df_output['in_event_type'] = df_DC["Transaction Details"].apply(event_type_map)
    
    # find the one with SSG0340 in it
    SSG_boo_list = df_output['in_event_type'] == Event_type['SSG0340']
    
    # find the SSG0340 corresponding Transaction Reference
    SSG = list(df_DC[SSG_boo_list]["Transaction Reference"])
    
    # check Transaction Reference with exactly 2 occurrance
    SSG_double_list = [k for k, v in list(Counter(SSG).items()) if v == 2]
    
    # filter out the one which is not in the above 2-occurrance list
    for idx in df_output[SSG_boo_list].index:
        if df_DC.loc[idx, 'Transaction Reference'] not in SSG_double_list:
            df_output.loc[idx, 'in_event_type'] = np.nan
    return df_output

df_output = in_event_type(df_output = df_output, df_DC = df_DC)
```

```python
# in_ccy

def in_ccy(inputStr):
    m = re.search(r'(?P<currency>\w+)$', inputStr)
    return m.groupdict()['currency']

df_output['in_ccy'] = df_DC['Account Name'].apply(in_ccy)
```

```python
# in_event_date
# Linedata not take weekend and public holiday(SG)

Pub_holiday=[
    '01/01/2019',
    '05/02/2019',
    '06/03/2019',
    '19/04/2019',
    '01/05/2019',
    '20/05/2019',
    '05/06/2019',
    '09/08/2019',
    '12/08/2019',
    '28/10/2019',
    '25/12/2019'
]

def get_bzday(inputDate, Pub_holiday = Pub_holiday):
    PH_dateobj = []
    for date in Pub_holiday:
        d,m,y = (int(i) for i in date.split('/'))
        PH_dateobj.append(datetime.datetime(y,m,d))

    day, month, year = inputDate.split('/')
    dateobj = datetime.datetime(int(year), int(month), int(day))

    while dateobj in PH_dateobj or dateobj.isoweekday()>5:    
        dateobj += datetime.timedelta(days=1)
        
    y,m,d = str(dateobj.year), str(dateobj.month), str(dateobj.day)
    return '/'.join([d,m,y])


df_output['in_event_date'] = df_DC['Business Date'].apply(get_bzday)
df_output['in_value_date'] = df_output['in_event_date']
```
