```python
import ast

def check_item_variation(data):
    for i in data['item_variation']:
        try:
            item_variation = ast.literal_eval(i)
            if len(item_variation.keys())>3:
                yield True
            else:
                yield False
        except:
            yield False
```

```python
flag = list(check_item_variation(data))
```

```python
len(data[flag]['itemid'].unique())
```

Use pandas function to identify duplicated listings within each shop (If listing A and B in shop S have the exactly same product title, product detailed description, and price, both listing A and B are considered as duplicated listings)
* a.	Mark those duplicated listings with True otherwise False and store the marking result in a new column named is_duplicated 
* b.	Find duplicate listings that has less than 2 sold count and store the result in a new excel file named duplicated_listings.xlsx
* c.	Find the preferred shop shopid that have the most number of duplicated listings

```python
data.head(3)
```

```python
columns
```

```python
is_duplicated = pd.Series()
for shop in list(data.groupby(['shopid'])):
    df_shopid = shop[1]
#     shopid = list(df_shopid['shopid'])[0]
    duplicate_df_shopid = df_shopid.duplicated(['itemid', 'item_description', 'price'])
#     print(duplicate_df_shopid)
    is_duplicated = is_duplicated.append(duplicate_df_shopid)
#     print("Duplicate:", type(duplicate_df_shopid), sep='\n')
```

```python
is_duplicated
```

```python
all(is_duplicated == False)

# There is no duplicate in this dataset
```

```python
df_is_duplicated = pd.DataFrame(data = is_duplicated, columns = ['is_duplicated'])
df_is_duplicated.head(3)
```

```python
df_data = df_is_duplicated.merge(data, left_index=True, right_index=True)
df_data.head(3)
```

```python
df_data_duplicate = df_data[df_data['is_duplicated'] == True]
df_data_duplicate
```
