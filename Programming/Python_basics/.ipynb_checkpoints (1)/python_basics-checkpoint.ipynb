# Python Tutorial: OS Module
##### allows us to interact with the underlying operating system in several different ways. eg. get file information, look up and change the environment variables, move files around and  all kinds of useful stuff. it's a build-in module

```python
import os
print(dir(os))
```

```python
#creat a folder on the desktop 'os-Demo-n'
# the first mothed
#os.mkdir('os-Demo-3') # seldom use it, can't create sub-dir
# second mothed
os.makedirs('os-Demo-2/Sub-Dir-1') # this one is better to create a tree function
```

```python
# remove a folder, similar use with mk & make
#os.rmdir('os-Demo-3') # will not remove the intermediate directories, removedirs will.
#when considering deleting directories recursively a little more dangerous than creating them
# so use rmdir is better coz you can specifically delete the exact directory
os.removedirs('os-Demo-2/Sub-Dir-1')
```

```python
print(os.listdir())
```

```python
#rename file name
os.rename('os-Demo-1','OS-Demo')
```

```python
# print out all information about 'OS-Demo'
print(os.stat('OS-Demo'))
```

```python
# print out all information about 'OS-Demo' with specific size use 'st_size'
print(os.stat('OS-Demo').st_size) 
```

```python
# print a human readable form of the last modification timestamp
from datetime import datetime 
mod_time = os.stat('OS-Demo').st_mtime
print(datetime.fromtimestamp(mod_time))
```

```python
# create a file, first consider the path
file_path = os.getcwd() + 'test.txt'
print(file_path)
# this method will bring confusion on the file name 
```

```python
# os.path module is good for working with paths, but here we use os.path.join
# to join two paths together
file_path2 = os.path.join(os.getcwd(), 'test1.txt')
print(file_path2)
```

```python
# grab the file name of any path that we are working on
# and this doesn't have to be a real path
print(os.path.basename('/tmp/test.txt'))
# get the directory name
print(os.path.dirname('/tmp/test.txt'))
# get both directory nae and basename 
print(os.path.split('/tmp/test.txt'))
```

```python
# check whether the file exists
print(os.path.exists('/tmp/test.txt'))
```

```python
# check it is a directory or a file
print(os.path.isdir('/tmp/test.txt'))
print(os.path.isfile('/tmp/test.txt'))
```

```python
# split the file root and file extension and get the file name without extension
print(os.path.splitext('/tmp/test.txt'))
```
