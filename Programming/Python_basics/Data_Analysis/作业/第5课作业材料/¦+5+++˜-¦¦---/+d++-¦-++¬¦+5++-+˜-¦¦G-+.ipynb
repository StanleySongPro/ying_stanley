5.通过dropna从df数据表中删除掉实习期限tcsl_duratin为空的数据行，删除掉Del内数值所在的数据行，并提取实习月份,要求数据格式为整数，不能是字符串，在df中创建名为“实习月数”的数据列。

```python
#删除掉实习期限tcsl_duratin为空的数据行
df=df.dropna(subset=['tcsl_duratin'])
```

```python
# 删除掉Del内数值所在的数据行
Del=[]
for i in df['tcsl_duratin'].index:
    if '&' in df['tcsl_duratin'][i] :
        Del.append(i)
df=df.drop(Del)
```

```python
df.index=range(len(df))
```

```python
df.loc[0]
```

```python
#提取实习月份
df['tcsl_duratin']
```

```python
intern=[i.replace('实习','') for i in df['tcsl_duratin']]
```

```python
intern=[i.replace('个月','') for i in intern]
```

```python
intern
```

```python
#数据格式为整数，不能是字符串
for i in range(len(intern)):
    intern[i] = int(intern[i])
intern 
```

```python
#在df中创建名为“实习月数”的数据列
df['实习月数']=intern
```
